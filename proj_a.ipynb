```python
import numpy as np

s_mean = []
est_var = []
s_var = []
n = 50
m = 200
mu = 5
sigma = 3
```

```python
for i in range(m):
    L = np.random.normal(mu, sigma, n)
    me = np.mean(L)
    s_mean.append(me)
    
    sig = 0
    for j in L:
        sig = sig + (j-me)*(j-me)
    sig = sig/n
    est_var.append(sig)
    
    s = (sig*n)/(n-1)
    s_var.append(s)
    print(L)
    print()
```

```python
print(np.round(s_mean,3))
print()
print(np.round(est_var,3))
print()
print(np.round(s_var,3))
```

```python
import pandas as pd
means = pd.DataFrame({'Mean distribution for n = 50':s_mean})
a = means.hist(bins = 40)
```

```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np


x_min = 3.0
x_max = 7.0

mean = 5.0
std = 0.42

x = np.linspace(x_min, x_max, 100)

y = scipy.stats.norm.pdf(x,mean,std)

plt.plot(x,y, color='blue')
plt.xlabel(x)
plt.xlim(x_min,x_max)
plt.ylim(0,1)
plt.title('Normal distribution with mean = 5 and sigma = 0.42')
plt.show()
```

```python
var = pd.DataFrame({'Variance distribution for n = 50':est_var})
var.hist(bins = 40)
```

```python
ssq = pd.DataFrame({'Sample variance (S square) distribution for n = 50':s_var})
ssq.hist(bins = 40)
```

```python
ssq2 = ssq
ssq2.rename(columns = {'Sample variance (S square) distribution for n = 50':'Comparison with chi-squared distribution'},inplace = True)
ssq2['Comparison with chi-squared distribution'] = ssq2['Comparison with chi-squared distribution']*((n-1)/(sigma*sigma)) 
ssq2.hist(bins = 40)
```

```python
bias_mean = [0]*len(s_mean)
bias_var = [0]*len(s_mean)
bias_svar = [0]*len(s_mean)

for i in range(len(s_mean)) :
    bias_mean[i] = abs(s_mean[i]-mu)
    bias_var[i] = abs(est_var[i]- sigma*sigma )
    bias_svar[i] = abs(s_var[i]-sigma*sigma)

    
avg_bias_mean = np.mean(bias_mean) 
avg_bias_var = np.mean(bias_var)
avg_bias_svar = np.mean(bias_svar)

print(avg_bias_mean)
print(avg_bias_var)
print(avg_bias_svar)
```

```python
avg_mean = np.mean(s_mean)
avg_var = np.mean(est_var)
avg_svar = np.mean(s_var)

var_mean = 0
var_var = 0
var_svar = 0

for i in range(len(s_mean)):
    var_mean += (s_mean[i]-avg_mean)*(s_mean[i]-avg_mean)
    var_var += (est_var[i]-avg_var)*(est_var[i]-avg_var)
    var_svar += (s_var[i]-avg_svar)*(s_var[i]-avg_svar)
    
var_mean = var_mean/(m-1)
var_var = var_var/(m-1)
var_svar = var_svar/(m-1)
```
